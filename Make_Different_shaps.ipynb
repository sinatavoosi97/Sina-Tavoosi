```python
#  Ø§ÛŒÙ† Ø¨Ø±Ù†Ø§Ù…Ù‡ Ù‚Ø§Ø¯Ø± Ø¨Ù‡ Ø³Ø§Ø®Øª Ø§Ø´Ú©Ø§Ù„ Ù…Ø®ØªÙ„ÙÛŒ Ù‡Ù…Ú†ÙˆÙ† Ù…Ø±Ø¨Ø¹ØŒÙ…Ø«Ù„Ø« Ù…ØªØ³Ø§ÙˆÛŒ Ø§Ù„Ø³Ø§Ù‚ÛŒÙ†ØŒÙ…Ø«Ù„Ø« Ù…ØªØ³Ø§ÙˆÛŒ Ø§Ù„Ø³Ø§Ù‚ÛŒÙ† Ø¨ØµÙˆØ±Øª Ù…Ø¹Ú©ÙˆØ³ Ùˆ Ù„ÙˆØ²ÛŒ
#                                                                                           .  Ù…ÛŒ Ø¨Ø§Ø´Ø¯

#=================decorating the terminal with the aim of showing better my code===========================
def detailing():
    line=(4*"ğ“€€ ğ“€ ğ“€‚ ğ“€ƒ ğ“€„ ğ“€… ğ“€† ğ“€‡ ğ“€ˆ ğ“€‰ ğ“€Š ğ“€‹ ğ“€Œ ğ“€ ğ“€ ")
    return line
#===================getting the length of shape from user==================================================
def get_length():
    length=int(input("please enter the length of shape  :"))
    return length
#===================getting the character of shape from user===============================================
def get_character():
    character=str(input("please enter the character of shape  :"))
    return character
#========This function makes a line with the goal of creating the shape that user want to make it=====================

def make_line(length,character):
    text=""
    for i in range(length):
        text+=character
    return text

def make_new_line():
    return "\n"
#======================= This function makes a square with user's information=============================================
def create_Square(length,character):
    text=""
    for i in range(length):
        text+=make_line(length,character)
        text+=make_new_line()
    return text

 #=======================This function makes a right angled triangle with user's information===========================================
def create_Right_angled_triangle(length,character):
    text=""
    for i in range(length):
        text+=make_line(((2*i)+1),character)
        text+=make_new_line()
    return text
#=======================This function makes a isosceles triangle with user's information=============================
def create_Isosceles_triangle(length,character):
    text=""
    for i in range(length):
        text+=make_line(length-i-1," ")
        text+=make_line(((2*i)+1),character)
        text+=make_new_line()
    return text

#=======================This function makes a reverse isosceles triangle with user's information=============================
def create_reverse_Isosceles_triangle(length,character):
    text=""
    count=(2*length)-3
    for i in range(length-1):
        text+=make_line(i+1," ")
        text+=make_line(count,character)
        count-=2
        text+=make_new_line()
    return text

#=======================This function makes a diamond with user's information===========================================
def create_Diamond(length,character):
    Diamond=create_Isosceles_triangle(length,character)+create_reverse_Isosceles_triangle(length,character)
    return Diamond


menu=("Square",
     "Right-angled triangle",
     "Isosceles triangle",
     "Reverse isosceles triangle",
     "Diamond",
    "Exit")
#==================================starting the program==================================================================
def main():
    while True:
        userchoice=show_menu()
        if userchoice==6:
            print("program exited successfully")
            break
        else:
            process(userchoice)
#================================showing the menu to user=======================================================    
def show_menu():  
    i=1
    for item in menu:
        message=(f"{i}-{item}")
        print(message)
        i+=1
    print(make_new_line())
    userchoice=int(input("please enter your choice :"))
    print(make_new_line())
    
    return userchoice
    
#=====================================processing on user's choices to make shapes=========================    
def process(userchoice): 
    
    if userchoice==1:
        print(make_new_line())
        length=get_length()
        character=get_character()
        print(make_new_line())
        print(create_Square(length ,character))
        print(detailing())
        
    elif userchoice==2:
        length=get_length()
        character=get_character()
        print(make_new_line())
        print(create_Right_angled_triangle(length,character))
        print(detailing())
    
    elif userchoice==3:
        length=get_length()
        character=get_character()
        print(make_new_line())
        print(create_Isosceles_triangle(length,character))
        print(detailing())
        
    elif userchoice==4:
        length=get_length()
        character=get_character()
        print(make_new_line())
        print(create_reverse_Isosceles_triangle(length,character))
        print(detailing())
    
    elif userchoice==5:
        length=get_length()
        character=get_character()
        print(make_new_line())
        print(create_Diamond(length,character))
        print(detailing())
        
           
    input("please press enter to continue .....")

main()
    
```
